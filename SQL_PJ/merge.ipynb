```python
import pandas as pd
```

```python
customer_info = pd.read_excel('1.1 고객정보.xlsx')
product_info = pd.read_excel('2.1 상품정보.xlsx')
order_info = pd.read_excel('3.1 주문정보.xlsx')
pay_info = pd.read_excel('4.1 결제정보.xlsx')
refund_info = pd.read_excel('5.1 환불정보.xlsx')
```
